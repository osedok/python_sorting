```python
# Adopted the source provided by https://www.programiz.com/dsa/insertion-sort

shirt_sizes=[10,5,15,3,7,4,8,21,6]
print(shirt_sizes)

def insertionSort(input_list):
    # Iterate through the input list elements
    for i in range(1, len(input_list)):
        element = input_list[i]
        prev_index = i-1
        # Now we compare key with each element 
        # on the left of it until an element smaller than it is found
        # The order of sorting can be changed to descending by changing:
        # element < input_list[prev_index] to element > input_list[prev_index]
        while prev_index >= 0 and element<input_list[prev_index]:
            input_list[prev_index+1] = input_list[prev_index]
            prev_index -= 1
           
        input_list[prev_index+1] = element
        
    return input_list


sorted_shirts = insertionSort(shirt_sizes)
print(sorted_shirts)
```

```python
# Quick sort in Python
# Adopted the source provided by https://www.programiz.com/dsa/quick-sort

students_heights=[165,192,181,172,154,166,180,201,168]
print("Unsorted Array:")
print(students_heights)
print("\nPartitioning: ")
print(students_heights)  

# Function to find the partition position
def partition (array, low, high, side):
    
 
    
  # Choose the rightmost element as a pivot
  pivot = array[high]
  
  # Pointer for greater element
  i = low - 1
  # Traverse through all elements and compare each element with the pivot
   

  if(side!="Input List"):
      print("Partitioning: ")
      print(array[low:high+1])

    
  for j in range (low, high):
    
    if array[j] <= pivot:
      # If an element smaller than the pivot is found
      # Swap it with the greater element pointed by i
      i = i + 1
      # Swapping element at i with the element at j
      (array[i], array[j]) = (array[j], array[i])

  # Swap the pivot element with the greater element specified by i
  (array [i + 1], array[high]) = (array[high], array [i + 1])
  
  pi = i+1

  print ("Selected pivot: "+ str(pivot))  
  print("Pivot moved to position: "+ str(pi))
  print(array)  
  print("\n")
 
  # Return the position from where the partition is done
  return pi

# Function to perform quicksort
def quickSort (array, low, high, side):
  if low < high:

    # Find a pivot element such that
    # Elements smaller than the pivot are on the left
    # Elements greater than the pivot are on the right
    pi = partition (array, low, high,side)
    # Recursive call on the left of the pivot
    quickSort (array, low, pi - 1, "Left")
    # Recursive call on the right of the pivot
    quickSort (array, pi + 1, high, "Right")
    return array




size = len(students_heights)
sorted_students_heights = quickSort(students_heights, 0, size - 1, "Input List")

print("\nSorted Array in Ascending Order:")
print(sorted_students_heights)
```

```python
# https://towardsdatascience.com/how-to-implement-merge-sort-algorithm-in-python-4662a89ae48c
def merge_lists(listA, listB):
    # Initialize an empty list to store the sorted elements
    # Initialize two variables i and j as pointers when iterating through the lists.
    listC = []
    i = j = 0

    # The while loop if both pointers i and j are less than the length of the left and right partitions
    while i < len(listA) and j < len(listB):
        # Compare the elements at every position of both lists during each iteration
        if listA[i] < listB[j]:
            # listC is populated with the smaller value
            listC.append(listA[i])
            # Move pointer i to the right
            i += 1
        else:
            listC.append(listB[j])
            # Move pointer j to the right
            j += 1
    ## The remnant elements are picked from the current pointer value to the end of the respective list
    listC.extend(listA[i:])
    listC.extend(listB[j:])

    return listC
```

```python
# https://towardsdatascience.com/how-to-implement-merge-sort-algorithm-in-python-4662a89ae48c
def merge_sort(input_list):
    
    # checking the size of the list
    size = len(input_list)
    
    # setting the base condition for the recuring function. 
    # if size of the list is = 1 -> it is a one element list and we will return it.
    if size ==1:
        return input_list
    
    # find the center index to split the input_list
    center = size // 2
            
    # recuring call for the partition on the left
    listA = merge_sort(input_list[:center])
    # recuring call for the partintion on the right
    listB = merge_sort(input_list[center:])
   
    #merge_lists
    return merge_lists(listA, listB)
   
```
